Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання

```python
import numpy as np
import pandas as pd
```

```python

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
tables = pd.read_html(url, match='Регіон', decimal=',', thousands='.')
data = tables[0]
data
```

Вивести перші рядки таблиці за допомогою методу head

```python
print(data.head())
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
shape = data.shape
print(f"рядків: {shape[0]}\nстовпців: {shape[1]}")
```

Замініть у таблиці значення "—" на значення NaN

```python
table = data.replace('—', np.nan)
table
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(table.dtypes)
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
data = table.apply(pd.to_numeric, errors='ignore')
print(data.dtypes)

```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
missing_percentage = (data.isnull().sum() / len(data) * 100).round(2)
missing_percentage
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
data = data.drop([data.shape[0]-1])
data
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
column_means = data.iloc[:, 1:].mean().round()
for column in data.columns[1:]:
    data[column].fillna(column_means[column], inplace=True)
data = data.round(2)
print(data)
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
regions = data[data['2019'] > data['2019'].mean()]['Регіон']
result = [region for region in regions]
result
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
region = data[data['2014'] == data['2014'].max()]['Регіон']
region.values[0]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
regions = data['Регіон']
year = data['2019']
plt.barh(regions, year)
plt.xlabel("Кількість", fontsize="small", color="midnightblue")
plt.ylabel("регіони", fontsize="small", color="midnightblue")
plt.title("народжуваність по регіонах у 2019 році", fontsize=10)
plt.show()
```

```python
regions = data['Регіон']
years = data.columns[1:]
data_numeric = data.iloc[:, 1:].apply(pd.to_numeric)
fig, axs = plt.subplots(figsize=(12, 8))
for i in range(len(regions)):
    axs.plot(years, data_numeric.iloc[i, :], label=regions[i])
axs.set_xlabel('рік')
axs.set_ylabel('народжуваність')
axs.set_title('зміна народжуваності в регіонах України (1950—2019)')
plt.xticks(rotation=45, ha='right')
axs.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
